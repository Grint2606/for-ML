```python
#Шаг 1: Загрузка набора данных из файла.

import pandas as pd
brainFrame = pd.read_csv("brainsize.txt", delimiter='\t') # Загружаем набор данных

brainFrame.head() # отображаем первые 5 строк

brainFrame.head(10) # отображаем первые 10 строк

brainFrame.tail(8) # отображаем последние 8 строк таблицы.
```

```python
brainFrame.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt #Импортируем библиотеки


menDf = brainFrame[brainFrame['Gender'] == 'Male']
womenDf = brainFrame[brainFrame['Gender'] == 'Female'] #Создаём два новых кадра данных

print(menDf)
print(womenDf)
```

```python
menMeanSmarts = menDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(menMeanSmarts, menDf["MRI_Count"])
plt.title('men diagram')
plt.xlabel('Mean Smarts')
plt.ylabel('MRI_Count')
plt.show()
```

```python
womenMeanSmarts = womenDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDf["MRI_Count"])
plt.title('Female diagram')
plt.xlabel('Mean Smarts')
plt.ylabel('MRI_Count')
plt.show()
```

```python
brainFrame_correal = brainFrame.corr(method='pearson')
print(brainFrame_correal) #корреляция для brainFrame.
```



Обратите внимание на диагональ слева направо в таблице корреляции, сгенерированной выше. Почему диагональ заполнена значениями 1? Это совпадение? Объясните.

(иагональ заполнена значениями 1 потому, что это корреляция между одинаковыми величинами, являющейся полной линейной зависимостью)






Продолжая смотреть на таблицу корреляции выше, обратите внимание, что значения зеркалируются; значения под диагональю имеют зеркальный аналог над ней. Это совпадение? Объясните.

(Значения под диагональю совпадают с зеркальными значениями над диагональю, потому что корреляция между переменными A и B такая же, как и корреляция между переменными B и A. Это свойство называют свойстом симметричности(зеркальности) матрицы корреляции.)



```python
womenDf_correal = womenDf.corr(method='pearson')
print(womenDf_correal) #корреляция переменных, содержащихся в кадре с женскими записями
```

```python
menDf_correal = menDf.corr(method='pearson')
print(menDf_correal) #корреляция переменных, содержащихся в кадре с мужскими записями
```



У многих пар переменных корреляция близка к нулю. Что это значит?

(если корреляция близка к нулю, это указывает на то, что изменения в одной переменной почти не влияют на изменения в другой переменной, Также это может означать, что между этими переменными низкая статистическая связь.)



Зачем делать разделение по полу?

(Разделение по полу позволяет проводить анализ и сравнение между группами мужчин и женщин. Это бывает полезным,из-за того что мужчины и женщины могут иметь различия в психологии, физиологии.)
